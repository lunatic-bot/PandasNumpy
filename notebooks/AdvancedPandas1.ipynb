```python
## Pivot tables
meanInfo = pd.pivot_table(salesData, index='Sub-Category', values=['Sales', 'Profit'], aggfunc='mean').reset_index()
meanInfo.columns = ['Sub-Category', 'MeanProfitVar', 'MeanSalesVar']
```

```python
## Pivot tables
stdInfo = pd.pivot_table(salesData, index='Sub-Category', values=['Sales', 'Profit'], aggfunc='std').reset_index()
stdInfo.columns = ['Sub-Category', 'StdProfitVar', 'StdSalesVar']
```

```python
save_the_table = pd.merge(meanInfo, stdInfo, on='Sub-Category', how='left')
save_the_table['diffStd'] = save_the_table['MeanSalesVar'] - save_the_table['StdSalesVar']
```

```python
save_the_table['diffMean'] = save_the_table['MeanProfitVar'] - save_the_table['StdProfitVar']
```

```python
save_the_table
```

```python
save_the_table.sort_values(by=['diffMean'], ascending=False)
```

## Group By

```python
salesData.groupby('Region').sum()['Profit']
```

```python
salesData.groupby(['Region', 'Customer ID']).sum().head(5)
```

```python
salesData.groupby(['Region', 'Customer ID']).agg({'Sales' : 'sum', 'Profit':'sum'}).reset_index()
```

```python
## two opeartion on same column - 
salesData.groupby(['Region', 'Customer ID']).agg(
    sumValSales=('Sales', 'sum'),
    varValSales=('Sales', 'var'),
    sumValProfit=('Profit', 'sum'),
    varValProfit=('Profit', 'var')
).reset_index()
```

```python
saveForEg = salesData.groupby(['Region', 'Customer ID']).agg({'Sales' : 'sum', 'Profit':'sum'}).reset_index()
saveForEg['ProfitRatio'] = saveForEg['Profit']/saveForEg['Sales']
```

```python
saveForEg = saveForEg.sort_values(['Region', 'ProfitRatio'], ascending=False).reset_index()
# saveForEg=saveForEg.re
saveForEg
```

```python
saveForEg.groupby(['Region']).agg({'ProfitRatio':{lambda x: x.nlargest(20).mean()}})
```

```python
saveForEg[saveForEg['Region'] == 'West'].head(20)['ProfitRatio'].mean()
```
